```python

import matplotlib.pyplot as plt
%matplotlib inline 
x=[12,3,36,45]
y=[1,2,4 ,45,]

p_x=[12,34,45,56,67,78]
p_y=[1,3,5,6,7,4]

#  Js Data 
j_x=[34,44,85]
j_y=[1,2,3]
```

```python


plt.xlabel('x axis')
plt.ylabel('Y-axis')
# plt.plot(x, y , linewidth=8 , linestyle="dashdot", color="Green", 'g+')
plt.plot(x,y  ,markersize=1 ,linewidth=2,  label="All developers")
plt.plot(p_x, p_y ,'rd--' ,markersize=1 ,linewidth=2,  label="Python Developers")
plt.plot(j_x, j_y ,'bd-.' ,markersize=1 ,linewidth=2,  label="Js Developers")
# plt.grid()
# plt.legend(['Js Developers' , 'Python Dev'])
plt.legend()

# plt.tight_layout()

# plt.style.use('seaborn')
plt.xkcd()
plt.savefig('test.png')

# print(plt.style.available)
# plt.legend('test')
```
